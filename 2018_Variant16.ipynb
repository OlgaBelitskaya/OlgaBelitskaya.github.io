### Задание 8 / Task 8
Площадь полной поверхности конуса равна 84.

Параллельно основанию конуса проведено сечение, делящее высоту пополам.

Найдите площадь полной поверхности отсеченного конуса.

---

The total cone surface area is equal to 84.

In parallel with the base of the cone, a section is made dividing the height in half.

Find the total area of the cut off cone.

```python
f = plt.figure(figsize=(7,7))
ax = f.add_subplot(111,projection='3d')

theta = numpy.linspace(0,2*numpy.pi,50)
r = numpy.linspace(0,4,50)
T, R = numpy.meshgrid(theta, r)

X = R * numpy.cos(T)
Y = R * numpy.sin(T)
Z = - numpy.sqrt(X**2 + Y**2)

ax.plot_wireframe(X, Y, Z, alpha=0.3)
Z[Z<-2] = numpy.nan
ax.plot_wireframe(X, Y, Z, colors='darkblue')

plt.axis('off');
```

Исходный конус подобен отсеченному с коэффициентом 2. 

Площади полных поверхности подобных тел относятся как квадраты коэффициента подобия.

The initial cone is similar to the cut off cone with the coefficient 2.

The areas of the total surfaces of similar solids are in the ratio of a square of the similarity coefficient.

$S_1 / S_2 = 2^2 \implies S_2 = S_1 / 4 = 84 / 4 = 21$
