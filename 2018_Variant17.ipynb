### Задание 8 / Task 8
Около шара описан цилиндр, площадь поверхности которого равна 15.

Найдите площадь поверхности шара.

---

A cylinder with a surface area of 15 is circumscribed around a sphere.

Find the surface area of the sphere.

```python
f = plt.figure(figsize=(9,9))
ax = f.add_subplot(111, projection='3d')

def create_sphere(center_x,center_y,center_z,radius):
    u, v = numpy.mgrid[0:2*numpy.pi:50j, 0:numpy.pi:50j]    
    x_grid = center_x + radius*numpy.cos(u)*numpy.sin(v)
    y_grid = center_x + radius*numpy.sin(u)*numpy.sin(v)
    z_grid = center_x + radius*numpy.cos(v)
    return x_grid,y_grid,z_grid

Xc,Yc,Zc = create_sphere(2.5,2.5,2.5,2.5)
ax.plot_surface(Xc, Yc, Zc, alpha=0.5, color='darkblue')

def create_cylinder(center_x,center_y,radius,height):
    z = numpy.linspace(0, height, 50)
    theta = numpy.linspace(0, 2*numpy.pi, 50)
    theta_grid, z_grid = numpy.meshgrid(theta, z)
    x_grid = radius * numpy.cos(theta_grid) + center_x
    y_grid = radius * numpy.sin(theta_grid) + center_y
    return x_grid,y_grid,z_grid

Xc,Yc,Zc = create_cylinder(2.5,2.5,2.5,5)
ax.plot_surface(Xc, Yc, Zc, alpha=0.3, color='steelblue')

plt.axis('off');
```

Радиусы шара и основания цилиндра равны $R$. Высота цилиндра равна $2 * R$.

The radii of the ball and the base of the cylinder are equal to $R$. The height of the cylinder is $2 * R$.

Площадь полной поверхности цилиндра равна сумме удвоенной площади основания и площади боковой поверхности:

The total surface area of the cylinder is equal to the sum of the doubled area of the base and the area of the lateral surface:

$S_{cylinder} = 2 * \pi * R^2 + 2 * \pi * R * 2 * R = 6 * \pi * R^2$

$S_{ball} = 4 * \pi * R^2 \implies S_{ball} \ / \ S_{cylinder} = 4 * \pi * R^2 \ / \ 6 * \pi * R^2 = 2 / 3$

$S_{ball} = 2 * S_{cylinder} \ / \ 3 = 2 * 15 / 3 = 10$

### Ответ / Answer: 10

```python
answer_dictionary[8] = str(10)
```
